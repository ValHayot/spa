```python
import numpy as np
def basic_model(xy1, xy2, xy3, xy4, num_pilots, xlabel, ylabel, system="Beluga", save=None):   
    from matplotlib import pyplot as plt
    fig, ax = plt.subplots()                                                     
    ax.scatter(xy1[0], xy1[1], c="purple", alpha=0.5)                                                  
    ax.scatter(xy2[0], xy2[1], c="red", alpha=0.5)
    ax.scatter(xy3[0], xy3[1], c="orange", alpha=0.5)
    ax.scatter(xy4[0], xy4[1], c="green", alpha=0.5)
    
    if len(xy1) > 2:
        ax.scatter(xy1[2], xy1[3], c="purple", marker="v", alpha=0.8)                                                  
        ax.scatter(xy2[2], xy2[3], c="red", marker="v", alpha=0.8)
        ax.scatter(xy3[2], xy3[3], c="orange", marker="v", alpha=0.8)
        ax.scatter(xy4[2], xy4[3], c="green", marker="v", alpha=0.8)
    
    get_c = lambda x, y, z=False: (10*(x+20)*125)/y if z else 10*(x+20)*np.ceil(125/y)
    ceil_results = False
    n_workers = np.arange(16, 70)
    
    a = 0.1
    ax.plot(n_workers, get_c(45, n_workers, ceil_results),'+', c='purple', alpha=a)
    ax.plot(n_workers, get_c(90, n_workers, ceil_results), '+', c='red', alpha=a)
    ax.plot(n_workers, get_c(120, n_workers, ceil_results), '+', c='orange', alpha=a)
    ax.plot(n_workers, get_c(180, n_workers, ceil_results), '+', c='green', alpha=a)
    
    
    #b, m = np.polynomial.polynomial.polyfit(x, y, 1)                             
    #ax.plot(np.asarray(x), b + m*np.asarray(x), 'k-')                            
    ax.set_xlabel(xlabel)                                                        
    ax.set_ylabel(ylabel)                                                        
                                                                                 
    if save is None:                                                             
        ax.set_title('{0} - {1} pilots'.format(system, num_pilots))              
        plt.show()                                                               
    else:                                                                        
        plt.savefig(save)  
```

```python
from statistics import mean, stdev
def pilot_queueing_times(submissions):
    return [[p['start_time'] - s['start_time']
             if p['start_time'] is not None else -1
             for p in s['sid']]for s in submissions if s['success']]

def avg_pilot_queueing(queue_times):
    result = {}
    
    for pilots in queue_times:
        for i in range(len(pilots)):
            if i in result:
                result[i].append(pilots[i])
            else:
                result[i] = [pilots[i]]
    return [(mean(v), stdev(v) if len(v) > 1 else 0) for k,v in sorted(result.items(), key=lambda x: x[0])]
```

```python
def fig_pilot_queue(pilot, batch, num_dedicated):
    import numpy as np
    import matplotlib.pyplot as plt
    from matplotlib.ticker import MaxNLocator
    from collections import namedtuple


    n_groups = len(pilot)

    mean_pilot = [el[0] for el in pilot]
    std_pilot = [el[1] for el in pilot]
    
    mean_pilot.append(0)
    std_pilot.append(0)
    
    mean_batch = [0 for el in mean_pilot]
    mean_batch[-1] = batch[0][0]

    std_batch = [0 for el in std_pilot]
    std_batch[-1] = batch[0][1]
    
    fig, ax = plt.subplots()

    index = np.arange(len(pilot) + 1)
    bar_width = 0.35

    opacity = 0.4
    error_config = {'ecolor': '0.3'}

    rects1 = ax.bar(index, mean_pilot, bar_width,
                    alpha=opacity, color='b',
                    yerr=std_pilot, error_kw=error_config,
                    label='pilots')

    rects2 = ax.bar(index, mean_batch, bar_width,
                    alpha=opacity, color='orange',
                    yerr=std_batch, error_kw=error_config,
                    label='batch')

    ax.set_xlabel('Pilot order')
    ax.set_ylabel('Average queueing time')
    ax.set_title('Average queueing time - {} dedicated'.format(num_dedicated))
    ax.set_xticks(index + bar_width / 2)
    ax.set_xticklabels((i for i in range(1, len(pilot) + 1)))
    ax.legend()

    fig.tight_layout()
    plt.show()
```
